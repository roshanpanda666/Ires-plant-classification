# ires plant classification

### data preprocessing

importing the libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

importing the dataset

```python
dataset=pd.read_csv('IRIS.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

splitting the dataset into train and test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
len(X_test)
```

```python
len(X_train)
```

```python
y
```

encoding the categorical data

```python
dataset.species.replace(('Iris-setosa', 'Iris-versicolor', 'Iris-virginica'), (1,2,3), inplace=True)
```

data preprocessing done

--------------------

### train the model

```python
from sklearn.svm import SVC
classifier = SVC(kernel = 'linear', random_state = 0)
classifier.fit(X_train, y_train)
```

here we are using SVM model

```python
dataset
```

```python
y_test  #original
```

```python
classifier.predict(X_test)  #predicted
```

### lets predict row number 4:- expected value( 1(ires-setosa) )

```python
print(classifier.predict([[5,3.6,1.4,0.2]])) #expected Iris-setosa
```

model is working properly

### lets see the accuracy

```python
classifier.score(X_test,y_test)
```

booom 97% accurate

### lets plot it and see how the classification seems like

```python
plt.plot(X_train,y_train,color='red')
plt.plot(X_test,y_test,color='blue')
plt.show()
```

red is original and blue is predicted

as we can see the threads are almost similar
